### Read shapefiles 

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np

# Read AGEBs'shapefile
buslines = "buslines/buslines.shp"
nbohood = "nbo_hood/nbo_hood.shp"

map_buslines = gpd.read_file(buslines)
map_buslines.describe()
map_buslines.head()

# map_nbohood = gpd.read_file(nbohood)
# map_nbohood.head()

```

### Adjust the coordinate system

```python
map_nbohood.crs 

# Change to Web Mercator projection
map_nbohood = map_nbohood.to_crs({'init': 'epsg:4326'}) # compatible with Google Maps
map_buslines = map_buslines.to_crs({'init': 'epsg:4326'})
```

```python
# Count how many polygons in each shapefile
map_buslines.count()
map_nbohood.count()

```

### Plot buslines

```python
f, ax = plt.subplots(1, figsize=(7, 7))
ax.set_title('Buslines Shapefile', fontdict={'fontsize': '20', 'fontweight' : '3'})
map_buslines.plot(facecolor='white', edgecolor='grey', ax=ax) 

plt.show()
```

### Plot neighborhoods' map

```python

f, ax = plt.subplots(1, figsize=(7, 7))
ax.set_title('Neighborhoords', fontdict={'fontsize': '20', 'fontweight' : '3'})

map_nbohood.plot(facecolor='white', edgecolor='grey', ax=ax) 

plt.show()
```

### Plot both maps

```python
f, ax = plt.subplots(figsize=(15, 20))
f.subplots_adjust(bottom=0.3)
ax.set_title('Portland neighborhoods and buslines', fontdict={'fontsize': '30', 'fontweight' : '3'})

map_nbohood.plot(facecolor='white', edgecolor='grey', linewidth=0.3, ax=ax) # NYNTA layer
map_buslines.plot(edgecolor='grey', linewidth=0.8, ax=ax) # neighborhoods' layer -- cloroperths with variable "random"

plt.show()
```

### Join both shapefiles

```python
nbohood_bus=gpd.sjoin(map_nbohood, map_buslines, op='intersects')
nbohood_bus

```

```python
output_data=nbohood_bus.drop(['geometry', 'AREA', 'LENGTH','FREQUENT'], axis=1)
output_data
```

```python
output_data.to_csv('Portland_NbohoodBuslines.csv')
```
